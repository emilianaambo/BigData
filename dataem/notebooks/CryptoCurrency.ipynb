Here is a heat map using correlation to see how the different prices of the different currencies change in comparison to each other.

```python


filesToUse = ["bitcoin_price.csv", "ethereum_price.csv", "ripple_price.csv", "litecoin_price.csv","dash_price.csv","monero_price.csv","nem_price.csv","neo_price.csv","numeraire_price.csv","omisego_price.csv","qtum_price.csv","stratis_price.csv","waves_price.csv"]

colsToUse = []
for ind, file_name in enumerate(filesToUse):
    currencyname = file_name.split("_")[0]
    if ind == 0:
        df = pd.read_csv("../data/raw/cryptocurrencypricehistory/"+file_name, usecols=["Date", "Close"], parse_dates=["Date"])
        df.columns = ["Date", currencyname]
    else:
        tempodf = pd.read_csv("../data/raw/cryptocurrencypricehistory/"+file_name, usecols=["Date", "Close"], parse_dates=["Date"])
        tempodf.columns = ["Date", currencyname]
        df = pd.merge(df, tempodf, on="Date")
    colsToUse.append(currencyname)
df.head()
        
tempodf = df[colsToUse]
corrmat = tempodf.corr(method='spearman')
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corrmat, vmax=1., square=True)
plt.title("Cryptocurrency correlation map", fontsize=15)
plt.show()
```
