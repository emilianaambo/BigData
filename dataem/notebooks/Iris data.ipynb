# Reading Iris Dataset 

```python
import numpy as np
import pandas as pd
```

```python
# reading the csv file 
dfiris = pd.read_csv('../data/raw/iris.data', header=None)
dfiris.head()
```

Here I am assigning column names to the dataset because there were no columns defined

```python
# Assigning column names 
dfiris.columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Species']
dfiris.head()
```

#### Visualize the Iris Data set

```python
# Pair plot 

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

plot = sns.pairplot(dfiris, vars=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'], hue='Species')

```

Here we can see from the visualizations of the pair plot that the Iris species setosa is separate from the two other species(versicolor and vi other two across all feature combinations



```python
# Plot using the violin plot
import seaborn as sns
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='Species',y='Sepal Length',data=dfiris)
plt.subplot(2,2,2)
sns.violinplot(x='Species',y='Sepal Width',data=dfiris)
plt.subplot(2,2,3)
sns.violinplot(x='Species',y='Petal Length',data=dfiris)
plt.subplot(2,2,4)
sns.violinplot(x='Species',y='Petal Width',data=dfiris)
```

```python
# plot using box and whisker 
dfiris.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()
```

```python
# plot using histograms
dfiris.hist(figsize=(15,20))
plt.figure()
```
