We have 3 different types of measures from the sensors, which are humidity, temperature and pressure and because of that I will group the 3 types into 3 seperate dataframes. In this case, I will concatinate all the Humidity data in one df, all temperature in another and so forth.  

```python
#concatinate dataframes
dfHumidity =  pd.concat([dfH_HTU21_measures, dfH_DHT22_measures, dfH_DHT11_measures], sort=True)
dfHumidity
```

```python
#converts column to datetime and removes the t and z characters
dfHumidity['time'] = pd.to_datetime(dfHumidity.time) 
dfHumidity
```

```python
#checks for any duplicates
dfHumidity.duplicated(subset=None, keep='first')
```

```python
#checks for any info of dataframe
dfHumidity.info()
```

```python
#count all data in each column
dfHumidity.count()
```

```python
#plots Humidity from the sensors
dfHumidity.plot.line(x='time', y='value')
```
